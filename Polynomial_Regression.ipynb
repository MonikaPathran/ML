```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
dataset=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv")
dataset
```

```python
x = dataset.iloc[:,1:2].values
x
```

```python
y =dataset.iloc[:,2].values
y
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg2 = PolynomialFeatures(degree=2)
x_poly = poly_reg2.fit_transform(x) #fit polynomial degree with 2
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg3 = PolynomialFeatures(degree=3)
x_poly3 = poly_reg3.fit_transform(x) #fit polynomial degree with 2
lin_reg3 = LinearRegression()
lin_reg3.fit(x_poly3,y)
```

```python
#polynomial regression
#degree rise
#linear polynomial
#predict
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title('LINEAR REGRESSTION')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x_poly3,lin_reg3.predict(x_poly3),color='blue')
plt.title('LINEAR REGRESSTION')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(poly_reg2.fit_transform(x)),color='blue')
plt.plot(x,lin_reg3.predict(poly_reg3.fit_transform(x)),color='green')
plt.title('LINEAR REGRESSTION')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()

```

```python
x = data.drop('Y house price of unit area',axis = 1)
x
```

```python
y = data['Y house price of unit area']
y
```

```python
polyregs = PolynomialFeatures(degree=2,include_bias=False)
```

poly_feat = polyregs.fit_transform(x)

```python
poly_feat = polyregs.fit_transform(x)
poly_feat.shape
```

```python
x_train,x_test,y_train,y_test= train_test_split(poly_feat,y,test_size=0.3,random_state=101)
```

```python
polymodel = LinearRegression()
```

```python
polymodel.fit(x_train,y_train)
```

```python
y_pred = polymodel.predict(x_test)
```

```python
pd.DataFrame({'y_test':y_test,'y_pred':y_pred,'Residuals':(y_test-y_pred)})
```

```python
poly_MAE = metrics.mean_absolute_error(y_test,y_pred)
poly_MSE = metrics.mean_squared_error(y_test,y_pred)
poly_RMSE = np.sqrt(poly_MSE)
```
